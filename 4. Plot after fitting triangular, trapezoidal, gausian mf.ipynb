Consider fuzzy set A with values in [1,10].
1. Fit triangular, trapezoidal and gaussean membership functions on this set.
2. Plot graphs for each membership function.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11, 0.1)


def triangular_mf(x, a, b, c):
  """
  Calculates the triangular membership function value.

  Args:
    x: The input value.
    a: The lower bound of the triangle.
    b: The peak of the triangle.
    c: The upper bound of the triangle.

  Returns:
    The membership value.
  """
  if x <= a or x >= c:
    return 0
  elif a <= x <= b:
    return (x - a) / (b - a)
  elif b <= x <= c:
    return (c - x) / (c - b)


def trapezoidal_mf(x, a, b, c, d):
  """
  Calculates the trapezoidal membership function value.

  Args:
    x: The input value.
    a: The lower bound of the left side.
    b: The start of the top side.
    c: The end of the top side.
    d: The upper bound of the right side.

  Returns:
    The membership value.
  """
  if x <= a or x >= d:
    return 0
  elif a <= x <= b:
    return (x - a) / (b - a)
  elif b <= x <= c:
    return 1
  elif c <= x <= d:
    return (d - x) / (d - c)


def gaussian_mf(x, mean, std):
  """
  Calculates the Gaussian membership function value.

  Args:
    x: The input value.
    mean: The mean of the Gaussian distribution.
    std: The standard deviation of the Gaussian distribution.

  Returns:
    The membership value.
  """
  return np.exp(-((x - mean) ** 2) / (2 * (std ** 2)))
```

```python
triangular_a = 2
triangular_b = 5
triangular_c = 8

trapezoidal_a = 1
trapezoidal_b = 3
trapezoidal_c = 7
trapezoidal_d = 9

gaussian_mean = 5
gaussian_std = 2


triangular_values = [triangular_mf(val, triangular_a, triangular_b, triangular_c) for val in x]
trapezoidal_values = [trapezoidal_mf(val, trapezoidal_a, trapezoidal_b, trapezoidal_c, trapezoidal_d) for val in x]
gaussian_values = [gaussian_mf(val, gaussian_mean, gaussian_std) for val in x]
```

```python
plt.plot(x, triangular_values, label='Triangular')
plt.plot(x, trapezoidal_values, label='Trapezoidal')
plt.plot(x, gaussian_values, label='Gaussian')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.title('Fuzzy Set Membership Functions')
plt.legend()
plt.show()
```
