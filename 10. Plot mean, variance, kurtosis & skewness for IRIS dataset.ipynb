Take the iris dataset and calculate the mean, variance, kurtosis & skewness. Plot the graph of mean, variance, kurtosis and skewness for the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.stats import kurtosis, skew

# Load the iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Calculate mean, variance, kurtosis, and skewness for each feature
mean_vals = df.mean()
variance_vals = df.var()
kurtosis_vals = df.apply(lambda x: kurtosis(x, fisher=True))
skewness_vals = df.apply(lambda x: skew(x))

# Combine all statistics into a single DataFrame for easy plotting
stats_df = pd.DataFrame({
    'Mean': mean_vals,
    'Variance': variance_vals,
    'Kurtosis': kurtosis_vals,
    'Skewness': skewness_vals
})

# Plotting the statistics
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Mean, Variance, Kurtosis, and Skewness of Iris Dataset Features')

# Mean plot
axs[0, 0].bar(stats_df.index, stats_df['Mean'], color='skyblue')
axs[0, 0].set_title('Mean')
axs[0, 0].set_xticklabels(stats_df.index, rotation=90)

# Variance plot
axs[0, 1].bar(stats_df.index, stats_df['Variance'], color='salmon')
axs[0, 1].set_title('Variance')
axs[0, 1].set_xticklabels(stats_df.index, rotation=90)

# Kurtosis plot
axs[1, 0].bar(stats_df.index, stats_df['Kurtosis'], color='lightgreen')
axs[1, 0].set_title('Kurtosis')
axs[1, 0].set_xticklabels(stats_df.index, rotation=90)

# Skewness plot
axs[1, 1].bar(stats_df.index, stats_df['Skewness'], color='orange')
axs[1, 1].set_title('Skewness')
axs[1, 1].set_xticklabels(stats_df.index, rotation=90)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
