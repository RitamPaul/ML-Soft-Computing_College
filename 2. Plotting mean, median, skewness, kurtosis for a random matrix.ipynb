# **1. Random (100x100) matrix, plot MEAN, MEDIAN, KURTOSIS, SKEWNESS**

```python
# importing dependencies

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
```

```python
row, col, minval, maxval = 100, 100, -10000, 10000

# creating matrix with random values

mat = np.random.randint(minval, maxval, size=(row,col))
print(mat)
```

```python
# creating empty arrays for mean and median

mean = np.empty(shape=col, dtype="float")
median = np.empty(shape=col, dtype="float")
sd = np.empty(shape=col, dtype="float")
skew = np.empty(shape=col, dtype="float")
kurt = np.empty(shape=col, dtype="float")

print("Mean = ")
print(mean)
print()
print("Median = ")
print(median)
print()
print("Standard Deviation = ")
print(sd)
print()
print("Skewness = ")
print(skew)
print()
print("Kurtosis = ")
print(kurt)
print()
```

```python
# filling the empty mean, median, skewness, kurtosis array

for j in range(0, col):
    # for mean
    li = []
    sum = 0
    for i in range(0, row):
        sum += mat[i][j]
        li.append(mat[i][j])

    mean[j] = sum / row

    # for median
    li.sort()
    left = math.floor((row+1) / 2)
    right = math.ceil((row+1) / 2)

    median[j] = (li[left-1] + li[right-1]) / 2

    # for standard deviation
    sdsum = 0
    for i in range(0, row):
        sdsum += (abs(mat[i][j] - mean[j]) ** 2)

    sd[j] = math.sqrt(sdsum / row)

    # for skewness
    skew[j] = 3 * (mean[j] - median[j]) / sd[j]

    # for kurtosis
    kurt[j] = kurtosis(li)

print("Mean = ")
print(mean)
print()
print("Median = ")
print(median)
print()
print("Standard Deviation = ")
print(sd)
print()
print("Skewness = ")
print(skew)
print()
print("Kurtosis = ")
print(kurt)
print()
```

```python
# plotting the MEAN points

plt.figure(figsize=(15,5))
plt.plot(mean, marker="o", color="r")
plt.title("MEAN graph")
```

```python
# plotting the MEDIAN points

plt.figure(figsize=(15,5))
plt.plot(median, marker="o", color="g")
plt.title("MEDIAN graph")
```

```python
# plotting the SKEWNESS points

plt.figure(figsize=(15,5))
plt.plot(skew, marker='o', color="b")
plt.title("SKEWNESS graph")
```

```python
# plotting the KURTOSIS points

plt.figure(figsize=(15,5))
plt.plot(kurt, marker='o', color="y")
plt.title("KURTOSIS graph")
```
