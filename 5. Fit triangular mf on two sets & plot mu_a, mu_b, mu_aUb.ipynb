Two fuzzy sets A and B contains 10 elements each. Fit triangular membership function on these sets. Combine the sets using fuzzy union and plot AuB, mu_AuB.

# OLD

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(0, 20, 0.1)


def triangular_mf(x, a, b, c):
  """
  Calculates the membership value for a triangular membership function.

  Args:
    x: The input value.
    a: The lower bound of the triangle.
    b: The peak of the triangle.
    c: The upper bound of the triangle.

  Returns:
    The membership value.
  """
  if x <= a or x >= c:
    return 0
  elif a < x <= b:
    return (x - a) / (b - a)
  elif b < x <= c:
    return (c - x) / (c - b)


a_a = 2
a_b = 5
a_c = 8
mu_a = [triangular_mf(i, a_a, a_b, a_c) for i in x]


b_a = 6
b_b = 10
b_c = 14
mu_b = [triangular_mf(i, b_a, b_b, b_c) for i in x]

mu_aub = [max(mu_a[i], mu_b[i]) for i in range(len(x))]
```

```python
plt.plot(x, mu_a, label='mu_A')
plt.plot(x, mu_b, label='mu_B')
plt.plot(x, mu_aub, label='mu_AuB')
plt.xlabel('x')
plt.ylabel('Membership Value')
plt.title('Fuzzy Union (AuB)')
plt.legend()
plt.grid(True)
plt.show()
```
