Consider 2 relations R and S where R(5x6) and S(6x3). Establish:
- max-min
- min-max
- max-product
- max-avg

On the composite relationship of R and S. Consider values of R and S random and between 0 and 1 (inclusive).

Also plot the graph of the 4 derived principles.

```python
import numpy as np
import matplotlib.pyplot as plt

R_rows, R_cols = 5, 6
S_rows, S_cols = 6, 3

np.random.seed(42)
R = np.random.rand(R_rows, R_cols)
S = np.random.rand(S_rows, S_cols)

def max_min_composite(R, S):
    return np.array([[np.max(np.minimum(R[i, :], S[:, j])) for j in range(S.shape[1])] for i in range(R.shape[0])])

def min_max_composite(R, S):
    return np.array([[np.min(np.maximum(R[i, :], S[:, j])) for j in range(S.shape[1])] for i in range(R.shape[0])])

def max_product_composite(R, S):
    return np.array([[np.max(R[i, :] * S[:, j]) for j in range(S.shape[1])] for i in range(R.shape[0])])

def max_avg_composite(R, S):
    return np.array([[np.mean(R[i, :] * S[:, j]) for j in range(S.shape[1])] for i in range(R.shape[0])])

composite_max_min = max_min_composite(R, S)
composite_min_max = min_max_composite(R, S)
composite_max_product = max_product_composite(R, S)
composite_max_avg = max_avg_composite(R, S)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Composite Relations of R and S")

# max-min
im = axs[0, 0].imshow(composite_max_min, cmap='viridis', aspect='auto')
axs[0, 0].set_title('Max-Min Composite')
fig.colorbar(im, ax=axs[0, 0])

# min-max
im = axs[0, 1].imshow(composite_min_max, cmap='viridis', aspect='auto')
axs[0, 1].set_title('Min-Max Composite')
fig.colorbar(im, ax=axs[0, 1])

# max-product
im = axs[1, 0].imshow(composite_max_product, cmap='viridis', aspect='auto')
axs[1, 0].set_title('Max-Product Composite')
fig.colorbar(im, ax=axs[1, 0])

# max-avg
im = axs[1, 1].imshow(composite_max_avg, cmap='viridis', aspect='auto')
axs[1, 1].set_title('Max-Avg Composite')
fig.colorbar(im, ax=axs[1, 1])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

composite_max_min, composite_min_max, composite_max_product, composite_max_avg
```
