```python
# importing dependencies

import random
import matplotlib.pyplot as plt
```

```python
n = int(input("Enter count of fuzzy member in set = "))

# creating arrays for member and membership function / grade

fmember = [100+member for member in range(1, n+1)]
fgradeA = [random.random() for i in range(n)]
fgradeB = [random.random() for i in range(n)]

print()
print("Fuzzy members = ")
print(fmember)
print()
print("Fuzzy set-A membership grades = ")
print(fgradeA)
print()
print("Fuzzy set-B membership grades = ")
print(fgradeB)
print()
```

```python
# performing fuzzy Union, Intersection

union, intersect = list(), list()

for i in range(n):
    union.append(max(fgradeA[i], fgradeB[i]))
    intersect.append(min(fgradeA[i], fgradeB[i]))

print("Fuzzy Set-A OR Fuzzy Set-B = ")
print(union)
print()
print("Fuzzy Set-A AND Fuzzy Set-B = ")
print(intersect)
print()
```

```python
# plotting UNION with all membership grades

plt.figure(figsize=(17,5))
plt.plot(fgradeA, marker="o", color="r", label="Membership Grade(Set-A)")
plt.plot(fgradeB, marker="o", color="g", label="Membership Grade(Set-B)")
plt.plot(union, marker="o", color="b", label="UNION(Set-A, Set-B)")
plt.legend(loc='upper right')
```

```python
# plotting INTERSECTION with all membership grades

plt.figure(figsize=(17,5))
plt.plot(fgradeA, marker="o", color="r", label="Membership Grade(Set-A)")
plt.plot(fgradeB, marker="o", color="g", label="Membership Grade(Set-B)")
plt.plot(intersect, marker="o", color="b", label="INTERSECTION(Set-A, Set-B)")
plt.legend(loc='upper right')
```
